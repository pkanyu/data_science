```python
# Classification in ML : It occurs in supervised learning when the target/outcome is categorical(textual)

# Step1 : Fetch Data
import pandas
data = pandas.read_csv('data/pima.csv')
```

```python
# Startistical analysis
data.describe()
```

```python
# Data Analysis and Preprocessing
```

```python
# Machine Learning
# Step 1: Splitting data into X and Y
matrix = data.values
matrix
```

```python
X = matrix[:,0:8]
Y = matrix[:,8]

Y
```

```python
# Splittind data into Training(70%) and Testing(30%) 
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
# Importing the scientific algorithms and Cross validating them(Pick the Best)
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# Pick one and create our model
model = GradientBoostingClassifier()
```

```python
# Training the model using the chosen algorithm
model.fit(X_train,Y_train)
```

```python
# Model Evaluation

predictions = model.predict(X_test)
predictions
```

```python
Y_test
```

```python
# Scientific Accuracy Score 
from sklearn.metrics import accuracy_score
accuracy_score(predictions,Y_test)
```
