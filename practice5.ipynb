```python
#Loading the data
import pandas
titles = pandas.read_csv('data/Movie_Id_Titles.csv')
titles.head(50)
```

```python
details = pandas.read_csv('data/file.csv')
details.head(50)
```

```python
data = pandas.merge(titles,details,on='item_id')
data.head(20)
```

```python
# Analyze
data.shape
```

```python
data.describe()
```

```python
#Lets see which title had a high rating on average
data.groupby('title')['rating'].mean().sort_values(ascending = False)
```

```python
#Lets see which title was viewed by majority of users
data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
# Combine the rating and the rates to a recommender system
ratings = pandas.DataFrame(data.groupby('title')['rating'].mean())
ratings['no_of_ratings'] = data.groupby('title')['rating'].count()
ratings.sort_values('no_of_ratings', ascending=False)


```
