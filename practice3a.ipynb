```python
#Get the data
import pandas
data = pandas.read_csv('data/school.csv')
data
```

```python
# Know the empties
data.isnull().sum()
```

```python
# One sample T test
# Fill in the empties
mean_math = data['Math'].mean()
data['Math'].fillna(mean_math,inplace=True)
```

```python
#Get the total mean
mean_population = data['Math'].mean()
mean_population
```

```python
#Get the sample
math_sample = data['Math'].sample(30)
```

```python
#Now we test the sample
from scipy.stats import ttest_1samp
statistics, pvalue = ttest_1samp(a=math_sample,popmean=mean_population)
print(pvalue)
if pvalue>0.05:
    print('Accept the Null Hypothesis')
    print("Reject the Alternative...")
else:
    print("Accept the Alternative")
    print('Reject the Null Hypothesis')
```
