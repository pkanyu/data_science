```python
#Recommender Systems:
#Ecommerce - product recommendation,Youtube - video recommendation, News App
#Two types of recommenders: 
#1. Collaborative Recommender : Recommendation is done based on other users reviews e,g trending video, trending hashtag
#2. Content Based -> Recommendation is done based on the users past activities and preferences e,g News Recommendation
```

```python
# Loading the dataset - IMDB(Movie Recommendation)
import pandas
titles = pandas.read_csv('data/Movie_Id_Titles.csv')
titles.head(50)
```

```python
details  = pandas.read_csv('data/file.csv')
details.head(6)
```

```python
data = pandas.merge(titles, details, on='item_id')
data.head(20)
```

```python
#Analysis: Shape(row, column)
data.shape
```

```python
# Statistical Analysis
data.describe()
```

```python
# Collaborative Filtering
# Highest rating 
# Majority of users

#Lets see which movie(title) was rated high(rating) on average
data.groupby('title')['rating'].mean().sort_values(ascending = False)
```

```python
#Lets see the movie that was viewed by majority of users
data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
# Combine rating and number of rating to a recommendation system
ratings = pandas.DataFrame(data.groupby('title')['rating'].mean())
ratings['no_of_ratings'] = data.groupby('title')['rating'].count()
ratings.sort_values('no_of_ratings', ascending=False)
```

```python
# Pivot Tables
# index - ydimension
# columns -xdime
# values- xdime

pivot = data.pivot_table(index='user_id',columns='title',values='rating')
pivot
```

```python
selected = pivot['Forrest Gump (1994)']
```

```python
similar = pivot.corrwith(selected)
similar_movie = pandas.DataFrame(similar, columns=['Correlations'])
similar_movie = similar_movie.join(ratings['no_of_ratings'])
similar_movie
```

```python
# Recommended Movie
similar_movie[similar_movie['no_of_ratings'] > 200].sort_values('Correlations', ascending=False).head(60)
```
