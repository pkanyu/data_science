```python
# Read the Exam Data
import pandas
data = pandas.read_csv('data/exams.csv')
data
```

```python
# HO: The mean of these 3 independent groups is the same
# H1: The mean is not the same
```

```python
from scipy.stats import f_oneway
statistic, pvalue = f_oneway(data['group1'],data['group2'],data['group3'])
print(pvalue)
if pvalue> 0.05:
    print('Accept the Null Hypothesis')
    print('Reject the Alternate ...')
else:
    print('Reject the Null Hypothesis ')
    print('Accept the  Alternate ...')
```
