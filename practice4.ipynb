```python
#Step 1
#Get Data
import pandas
data = pandas.read_csv('data/Advertising.csv')
data
```

```python
#Step two: Clean data
#Step three: Data Wrangling(We transform the data to the needed values)
#Step Four: Data Analysis and visualization
import matplotlib.pyplot as plt
plt.xlabel('Sales')
plt.ylabel('TV')
plt.title("The Sales against TV")
plt.scatter(data['TV'],data['Sales'])
plt.show()
```

```python
#Step Five,Split the data into matrix and Multi-dimensional arrays
matrix = data.values

X = matrix[:,0:3]
Y = matrix[:,3]
data.values
```

```python
#Step Six, split your data into train and test
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y,test_size=0.3,random_state=42)

```

```python
# Step Seven , make the necessary imports
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
```

```python
# Step Eight, Model training through Cross Validation
model = DecisionTreeRegressor()
model.fit(X_train,Y_train)
```

```python
# Step Nine, Our model has now been trained and now it is time to test it ,the X_test
predictions = model.predict(X_test)
print(predictions)
```

```python
# Step Ten, Againt the Y_test
print(Y_test)
```

```python
# Step Eleven, Now let us group our findings together
from sklearn.metrics import r2_score
print(f'Our Accuracy is {r2_score(predictions,Y_test)}')
```

```python
# Step Twelve,Now you are ready to make predictions on your own using your model
adverts = [[300,210,0]]
sales = model.predict(adverts)
print(sales)
```
